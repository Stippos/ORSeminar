```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import tree

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# Load the data
df = pd.read_csv("/Users/Max/Dropbox/Kevät 2018/McKinsey/pyt/data_set_continuous.csv") # For my Macbook
# df = pd.read_csv("...") # For my Windows PC
df.head(5)
```

```python
# Prepare data for training
features = ['Sex','Education','Married','Age_categorical','Balance_categorical','Location2']
X = df[features].copy()
y = df['Default'].copy()
```

```python
# Let's split the data by first splitting it into two groups (24000 and 6000) then divide the first split into two
testsize = (6/30)
random = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=testsize, random_state=random)
```

```python
# Split the 24000 into two
half = int(len(X_train)/2)
fullLength = int(len(X_train))
split1 = list(range(0,half))
split2 = list(range(half,fullLength))

# Define training sets 1 and 2
# The third set of 6000 is x_test,y_test
X_train_1 = X_train.iloc[split1]
X_train_2 = X_train.iloc[split2]
y_train_1 = y_train.iloc[split1]
y_train_2 = y_train.iloc[split2]
# Check sizes
print('X_train 1, X_train_2 lengths:')
print(len(X_train_1),len(X_train_2))
print('y_train_1, y_train_2 lengths:')
print(len(y_train_1),len(y_train_2))
print('X_test, y_test lengths:')
print(len(X_test),len(y_test))
```

```python
# Train the decision tree
dt = tree.DecisionTreeClassifier(criterion='entropy',max_depth=8,min_samples_leaf=100)
dt.fit(X_train_1,y_train_1)
```

```python
# Let's gather only the end nodes that have a certain probability of default in them
condition = 0.51
rows = 30000
rowNumbers = []
listOfLeafs = []
probabilities = []
for row in range(0,rows):
    sample = np.array(X.iloc[row,])
    defaultRatio = dt.predict_proba(sample.reshape(1,-1))[0,1]
    if defaultRatio > condition:
        probabilities.append(defaultRatio)
        endLeaf = int(dt.apply(sample.reshape(1,-1)))
        listOfLeafs.append(endLeaf)
        rowNumbers.append(row)
```

```python
print(len(set(listOfLeafs)),len(listOfLeafs))
print(set(listOfLeafs))
```

```python
highprobdf = df.iloc[rowNumbers,]
highprobdf['Leaf'] = listOfLeafs
highprobdf.head(5)
```

## Create a pd dataframe from the end nodes that satisfy our probability condition

### This way we can interpret what the segments look like

```python
columnNames = []
for element in list(set(listOfLeafs)):
    string = str(element)
    columnNames.append(string)
print(columnNames)
```

```python
# Creating the data structure as arrays
varNames = ['Sex', 'Education','Married','Age_categorical','Balance_categorical','Location2','Default']
N = len(set(listOfLeafs))
d = np.zeros([N*7,6]).astype(int)
i = 0
for leaf in set(listOfLeafs):
    currentLeafFrame = highprobdf[highprobdf['Leaf']==leaf]
    print('Leaf:',leaf)
    print('Samples in leaf: ',len(currentLeafFrame))
    for variable in varNames:
        uniques = currentLeafFrame[variable].unique().tolist()
        for varValue in uniques:
            count = len(currentLeafFrame[currentLeafFrame[variable]==varValue])
            print(variable,' == ',varValue,':', count)
            d[i,varValue] = count
        i = i + 1
    print('---------')
    
# Multi-index dataframe will suit our purpose
pdIndex = pd.MultiIndex.from_product([columnNames,varNames])
treeFrame = pd.DataFrame(data=d,index=pdIndex)
```

```python
treeFrame
```

```python
i = 0
defaultRatios = [None]*len(columnNames)
samplesInLeaf = [None]*len(columnNames)
for leaf in columnNames:
    subTree = treeFrame.loc[leaf].loc['Default']
    samplesInLeaf[i] = sum(subTree)
    defaultRatios[i] = subTree[1] / samplesInLeaf[i]
    i = i + 1
defaultRatios = np.array(defaultRatios)
defaultRatios = np.round_(defaultRatios,decimals=2)
print('Leafs and the corresponding default ratios and number of samples in them')
print(np.array([list(set(listOfLeafs)),defaultRatios]))
print(samplesInLeaf)
print('Total number of samples in the leafs:', sum(samplesInLeaf))
```
